```python
df2 = pd.read_csv("./data/소비자_물가지수.csv", encoding="euc-kr").dropna()
df2 = df2.transpose().reset_index(drop=True)
df2 = df2.rename(columns=df2.iloc[0])
df2 = df2.iloc[1:]
df2 = df2.rename(columns={"국가별": 'year'})
df2['year'] = df2['year'].astype(int)
df2 = df2.set_index(keys="year")
df2
```

```python
df2 = df2[df2.loc[:, :] != '-']
df2 = df2.astype(float)
df2 = df2.fillna(0)
df2
```

```python
snp = pd.read_csv("./data/s&p500.csv", encoding="euc-kr")
nifty = pd.read_csv("./data/nifty.csv", encoding="euc-kr")
nasdaq = pd.read_csv("./data/nasdaq.csv", encoding="euc-kr")
kospi = pd.read_csv("./data/kospi.csv", encoding="euc-kr")
hs = pd.read_csv("./data/hangseng.csv", encoding="euc-kr")
ftse = pd.read_csv("./data/ftse.csv", encoding="euc-kr")
dow = pd.read_csv("./data/dowjones.csv", encoding="euc-kr")
dax = pd.read_csv("./data/dax.csv", encoding="euc-kr")
asnp = pd.read_csv("./data/austs&p.csv", encoding="euc-kr")
nikkei = pd.read_csv("./data/nikkei.csv", encoding="euc-kr")

snp = snp.sort_values(by="날짜").reset_index(drop=True)
nifty = nifty.sort_values(by="날짜").reset_index(drop=True)
nasdaq = nasdaq.sort_values(by="날짜").reset_index(drop=True)
kospi = kospi.sort_values(by="날짜").reset_index(drop=True)
hs = hs.sort_values(by="날짜").reset_index(drop=True)
ftse = ftse.sort_values(by="날짜").reset_index(drop=True)
dow = dow.sort_values(by="날짜").reset_index(drop=True)
dax = dax.sort_values(by="날짜").reset_index(drop=True)
asnp = asnp.sort_values(by="날짜").reset_index(drop=True)
nikkei = nikkei.sort_values(by="날짜").reset_index(drop=True)
```

```python
snp = snp.iloc[:, :2]
snp = snp.rename(columns = {'종가': 's&p500'})
nifty = nifty.iloc[:, :2]
nifty = nifty.rename(columns = {'종가': 'nifty'})
nasdaq = nasdaq.iloc[:, :2]
nasdaq = nasdaq.rename(columns = {'종가': 'nasdaq'})
kospi = kospi.iloc[:, :2]
kospi = kospi.rename(columns = {'종가': 'kospi'})
hs = hs.iloc[:, :2]
hs = hs.rename(columns = {'종가': 'hangseng'})
ftse = ftse.iloc[:, :2]
ftse = ftse.rename(columns = {'종가': 'ftse'})
dow = dow.iloc[:, :2]
dow = dow.rename(columns = {'종가': 'dowjones'})
dax = dax.iloc[:, :2]
dax = dax.rename(columns = {'종가': 'dax'})
asnp = asnp.iloc[:, :2]
asnp = asnp.rename(columns = {'종가': 'aust_s&p'})
nikkei = nikkei.iloc[:, :2]
nikkei = nikkei.rename(columns = {'종가': 'nikkei'})
```

```python
indexes = [snp, nifty, nasdaq, kospi, hs, ftse, dow, dax, asnp, nikkei]


for i in range(1, len(indexes)):
    if(i == 1):
        economic_index = pd.merge(left=snp, right=nifty, how='outer', on='날짜')
    else:
        economic_index = pd.merge(left=economic_index, right=indexes[i], how='outer', on='날짜')

economic_index = economic_index.sort_values("날짜").reset_index(drop=True)
economic_index
```

```python
economic_index['날짜'] = pd.to_datetime(economic_index['날짜'])
for i in indexes:
    economic_index[i.columns[1]] = economic_index[i.columns[1]].str.replace(",", "").astype("float64")
```

```python
economic_index["year"] = economic_index['날짜'].dt.year
economic_index
```

```python
year_index_mean = economic_index[["year", "s&p500", "nifty", "nasdaq", "kospi", "hangseng", "ftse", "dowjones", "dax", "aust_s&p", "nikkei"]].groupby(['year'], as_index=False).mean()
```

```python
year_index_mean
```

```python
price_index_table = pd.merge(left=df2, right=year_index_mean, how='inner', on="year")
price_index_table = price_index_table.dropna()
price_index_table = price_index_table.set_index(keys="year")
price_index_table
```

```python
country = '이스라엘'

scaler1 = preprocessing.MinMaxScaler()
scaler2 = preprocessing.MinMaxScaler()

country_X = price_index_table
country_y = price_index_table[[country]]

scaler_x = scaler1.fit(country_X)
scaler_y = scaler2.fit(country_y)

X_train = scaler_x.transform(country_X[country_X.index<2019])
X_test = scaler_x.transform(country_X[country_X.index>=2019])
y_train = scaler_y.transform(country_y[country_y.index<2019])
y_test = scaler_y.transform(country_y[country_y.index>=2019])
```

### 선형회귀 모델

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)

print("가중치(계수, 기울기 파라미터 W) :", model.coef_)
print("편향(절편 파라미터 b) :", model.intercept_)

print("훈련세트 점수: {:.2f}".format( model.score(X_train, y_train) ))
print("테스트세트 점수: {:.2f}".format( model.score(X_test, y_test) ))
```

```python
y_predicted = model.predict(X_test).reshape([-1, 1])
pd.DataFrame(scaler_y.inverse_transform(y_predicted), columns=['result'])
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure(figsize=(8,12))

  plt.subplot(2,1,1)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
#   plt.ylim([0,5])
  plt.legend()

  plt.subplot(2,1,2)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  # plt.ylim([0,1])
  plt.legend()
  plt.show()

plot_history(history)
```

```python
years = list(price_index_table.index)

for i in range(len(years)):
    years[i] = datetime.datetime.strptime(str(years[i])+"/01/01", "%Y/%m/%d")

print(type(years[0].date()))

price_index_table.index = years
```

```python
dates = pd.to_datetime(price_index_table.index)
cols = list(price_index_table)[190:]
```

```python
country = '대한민국'
cols.append(country)

country_data = price_index_table[cols].astype(float)
```

```python
country_data
```

```python
scaler = StandardScaler()
scaler = scaler.fit(country_data)
country_data_scaled = scaler.transform(country_data)
```

```python
country_data_scaled.shape
```

```python
n_train = int(0.9*country_data_scaled.shape[0])
train_data_scaled = country_data_scaled[0: n_train]
train_dates = dates[0: n_train]

test_data_scaled = country_data_scaled[n_train:]
test_dates = dates[n_train:]
```

```python
trainX = train_data_scaled
trainY = train_data_scaled[:, -1]
testX = test_data_scaled
testY = test_data_scaled[:, -1]

# for i in range(seq_len, n_train-pred_days +1):
#     print(trainX)
#     trainX.append(train_data_scaled[i - seq_len:i, 0:train_data_scaled.shape[1]])
#     trainY.append(train_data_scaled[i + pred_days - 1:i + pred_days, 0])

# for i in range(seq_len, len(test_data_scaled)-pred_days +1):
#     testX.append(test_data_scaled[i - seq_len:i, 0:test_data_scaled.shape[1]])
#     testY.append(test_data_scaled[i + pred_days - 1:i + pred_days, 0])

trainX, trainY = np.array(trainX), np.array(trainY)
# trainX  = trainX.reshape(trainX.shape[0], trainX.shape[1], 1)
testX, testY = np.array(testX), np.array(testY)
# testX = testX.reshape(testX.shape[0], testX.shape[1], 1)
```

```python
trainY.shape
```

```python
# LSTM 모델 정의
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(trainX.shape[1], 1)))
model.add(Dense(trainY.shape[0]))  # 출력 뉴런 수는 출력 열의 수와 동일하게 설정
model.compile(optimizer='adam', loss='mse')

# 모델 학습
model.fit(trainX, trainY, epochs=100, batch_size=1)

# 예측을 위한 데이터 포맷 변환
# 예측
predicted_value = model.predict(testX)

# 예측 결과 역 스케일링
predicted_value = scaler.inverse_transform(np.reshape(predicted_value, (predicted_value.shape[1], -1)))

print(f"다음 물가지수 예측값: {predicted_value[0]}")
```
