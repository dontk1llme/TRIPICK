```python
selected_order = ['year'] + list(price_index_table.columns[4:])

country = price_index_table[price_index_table['country'] == '일본']
country['price_index'] = country['price_index'].astype("float64")
country
```

```python
country_X = country[['year', 's&p500', 'nifty', 'nasdaq', 'kospi', 'hangseng', 'ftse', 'dowjones', 'dax', 'aust_s&p', 'nikkei']]
country_y = country[['year', 'price_index']]

X_train = country_X[country_X['year']<2019]
X_test = country_X[country_X['year']>=2019]
y_train = country_y[country_y['year']<2019]['price_index']
y_test = country_y[country_y['year']>=2019]['price_index']
```

```python
from sklearn.model_selection import cross_val_score
```

### 선형회귀 모델

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

print("가중치(계수, 기울기 파라미터 W) :", model.coef_)
print("편향(절편 파라미터 b) :", model.intercept_)

print("훈련세트 점수: {:.2f}".format( model.score(X_train, y_train) ))
print("테스트세트 점수: {:.2f}".format( model.score(X_test, y_test) ))
```

```python
pd.DataFrame(model.predict(X_test), columns=['result'])
```

### Random Forest 모델

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=3000, max_depth=7000)
model.fit(X_train, y_train)

print("훈련세트 점수: {:.2f}".format( model.score(X_train, y_train) ))
print("테스트세트 점수: {:.2f}".format( model.score(X_test, y_test) ))

pd.DataFrame(model.predict(X_test), columns=['result'])
```

### DecisionTreeRegressor 모델

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=0)
model.fit(X_train, y_train)

print("훈련세트 점수: {:.2f}".format( model.score(X_train, y_train) ))
print("테스트세트 점수: {:.2f}".format( model.score(X_test, y_test) ))

pd.DataFrame(model.predict(X_test), columns=['result'])
```

### GradientBoostingRegressor 모델

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

### Sequential 모델

```python
model = keras.Sequential()
model.add(layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32))

optimizer = tf.keras.optimizers.RMSprop(0.001)

model.compile(loss="mse", optimizer=optimizer, metrics=['mse', 'mae'])
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=1000, validation_split=0.2, verbose=0)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure(figsize=(8,12))

  plt.subplot(2,1,1)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
#   plt.ylim([0,5])
  plt.legend()

  plt.subplot(2,1,2)
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
#   plt.ylim([0,20])
  plt.legend()
  plt.show()

plot_history(history)
```
